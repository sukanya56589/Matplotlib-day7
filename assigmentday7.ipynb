```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Chart

x = [2, 4, 6, 8]
y = [5, 15, 25, 35]

plt.plot(x, y, color='blue', marker='s', linestyle='--')

plt.title("Student Performance Chart")
plt.xlabel("Tests")
plt.ylabel("Marks")

plt.show()
```

```python
# Bar Chart

subjects = ['Math', 'Science', 'English', 'Social']
marks = [85, 90, 78, 88]

plt.bar(subjects, marks, color='skyblue')

plt.title("Subject Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()
```

```python
# Histogram

data = [5, 7, 8, 10, 12, 12, 14, 15, 17,
        18, 19, 20, 21, 22, 24, 25, 27]

plt.hist(data, bins=5, color='pink')

plt.title("Histogram Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
# Scatter Plot

x = [1, 2, 3, 4, 5]
y = [9, 14, 18, 24, 30]

plt.scatter(x, y, color='darkgreen')

plt.title("Scatter Plot Graph")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```

```python
# Pie Chart

games = ['Cricket', 'Football', 'Chess', 'Tennis']
students = [40, 25, 20, 15]

plt.pie(students, labels=games, autopct='%1.1f%%')

plt.title("Sports Interest")

plt.show()
```

```python
# Box Plot

data = [
    [3, 5, 7, 8, 10],
    [11, 13, 15, 16, 18],
    [20, 21, 23, 25, 27]
]

plt.boxplot(data)

plt.title("Student Data Box Plot")
plt.xlabel("Groups")
plt.ylabel("Values")

plt.show()
```

```python
# Heatmap

np.random.seed(3)

data = np.random.rand(6, 6)

plt.imshow(data, cmap='plasma')

plt.colorbar()

plt.title("Heatmap Visualization")

plt.show()
```
